```python
# Your code here ...

def make_list_int(my_list):
    new_list = []
    for i in my_list:
        tmp = int(i)
        new_list.append(tmp)
    return new_list

def get_avg_price(z, infile):
    f_open = open(infile)
    l_remove_newline = ''
    price_list = []
    for l in iter(f_open):
        l_remove_newline = l.rstrip()
        split_l = l_remove_newline.split(",")
        iter_zip = split_l[8]
        if (iter_zip == z):
            price_list.append(split_l[9])
        else:
            pass
    int_price_list = make_list_int(price_list)
    avg = sum(int_price_list) / len(price_list)
    return avg
    
    
def get_lists(filename):
    infile = filename
    file_open = open(filename)
    ZipList = []       # This is a list of all evaluated zipcodes
    AvgPrice_list = []
    AvgPrice4Zip = 0
    cur_zip = ''      # This is the zip whose average home price is being evaluated
    for line in iter(file_open):
        line_no_newline = line.rstrip()
        line_split = line_no_newline.split(",")
        cur_zip = line_split[8]
#        print("curzip =", cur_zip, "ziplist =", ZipList)
        if cur_zip in ZipList:
            pass
        else:
            AvgPrice4Zip = get_avg_price(cur_zip, infile)
            AvgPrice_list.append(AvgPrice4Zip)
            ZipList.append(line_split[8])
#            print("avg_price", AvgPrice_list, "ziplist", ZipList)
    file_open.close()
    return ZipList,AvgPrice_list

def get_top_50(a):
    api_50 = []
    count = 0
    for line in iter(a):
        api_50.append(line[1])
        count +=  1
        if count == 50 :
            return api_50

x,y = get_lists('Bayarea_homes.csv')   # This generates two lists of zipcodes (x) and avg homeprices(y)
print(x)
print(y)

fig, ax = plt.subplots(figsize=(5,5))
plt.title(" BAR CHART of ZIPCODE VS AVG HOME PRICE")
plt.xlabel("Zip Code")
plt.ylabel("Average Home Price")
ax.bar(x,y)
```
